```python
import pandas as pd
import json
import os
```

```python
with open(".\data\Streaming_History_Audio_2018_4.json", encoding="utf8") as f:
    a = json.load(f)
```

```python
df = pd.DataFrame(a)
```

```python
df[df["ms_played"]==0].groupby("master_metadata_track_name").count().sort_values(by = "ts", ascending=False)
```
